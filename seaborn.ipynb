```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tips = pd.read_csv('./tips.csv')
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='time')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='sex',style='time',size='size',kind='scatter')
```

```python
gap= pd.read_csv('./gap.csv')
gap
```

```python
temp_df = gap[gap['country']=='India']
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line')
```

```python
temp_df = gap[gap['country'].isin(['India','Brazil','Germany'])]
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line',hue='country')
```

```python
sns.relplot(data=gap,x='lifeExp',y='gdpPercap',kind='line',hue='country',col_wrap=3,col='continent')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='sex')
```

```python
sns.displot(data=tips,x='total_bill',kind='hist')
```

```python
sns.histplot(data=tips,x='total_bill',kde=True)
```

```python
sns.displot(data=tips,x='total_bill',kind='hist',hue='sex',element='step')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True)
```

```python
# rug plot
sns.displot(data=tips,x='total_bill',kind='kde',rug=True)
```

```python
gap=gap.pivot(index='country',columns='year',values='lifeExp')
sns.heatmap(gap)
```
