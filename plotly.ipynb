```python
import numpy as np
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go
avg = pd.read_csv('./avg.csv')
avg.head()
avg = avg.rename(columns={"batsman_runs":"SR"})
avg.head(1)
```

```python

traces = go.Scatter(x = avg['avg'],y=avg['SR'],mode = 'markers')
data = [traces]
layout = go.Layout(title = 'Batsman Avg vs SR',xaxis ={'title' :'Batsman Average'},yaxis = {'title' : 'Batsman Strike Rate'})
fig = go.Figure(data=data, layout=layout)
pyo.plot(fig)
```

```python
ipl = pd.read_csv('./ipl.csv')
single = ipl[ipl['batsman'] == 'V Kohli']
performance = single.groupby('season')['batsman_runs'].sum().reset_index()
performance
```

```python
traces = go.Scatter(x = performance['season'],y=performance['batsman_runs'],mode = 'lines+markers',marker={'color':'red'})
data = [traces]
layout = go.Layout(title = 'Kohli Performance',xaxis ={'title' :'Season'},yaxis = {'title' : 'Runs'})
fig = go.Figure(data=data, layout=layout)
pyo.plot(fig)
```

```python
single1=ipl[ipl['batsman']=='V Kohli']
performance1=single1.groupby('season')['batsman_runs'].sum().reset_index()
performance1

single2 = ipl[ipl['batsman'] == 'MS Dhoni']
performance2 = single2.groupby('season')['batsman_runs'].sum().reset_index()
performance2
```

```python
trace=go.Scatter(x=performance1['season'],y=performance1['batsman_runs'],mode='lines+markers',name='V Kohli')
trace1=go.Scatter(x=performance2['season'],y=performance2['batsman_runs'],mode='lines+markers',name='MS Dhoni')
data=[trace,trace1]
layout=go.Layout(title='Kohli vs Dhoni',xaxis={'title':'Season'},yaxis={'title':'Runs'})
fig=go.Figure(data=data,layout=layout)
pyo.plot(fig)
```

```python
df['family_size'] = df['SibSp']+df['Parch']+1
df
```

```python
def transform_family_size(size):
    if size == 1:
        return 'Alone'
    elif size <= 4:
        return 'Small'
    else:
        return 'Large'  
```

```python
df['family_type'] = df['family_size'].apply(transform_family_size)
df
```

```python
# calculate mean , mode , median of [40,20,20,60,80,60,20,60] using numpy
data = [40,20,20,60,80,60,20,60]
mean = np.mean(data)
median = np.median(data)
print(mean,median)
# find the mode of [40,20,20,60,80,60,20,60] using numpy
from scipy import stats
mode = stats.mode(data, keepdims=True)
print(mode)
```

```python
# draw box plot for inter quartile range of [40,20,20,60,80,60,20,60] using numpy
import matplotlib.pyplot as plt
plt.boxplot(data)
plt.show()
```

```python
import numpy as np
from scipy.stats import iqr

data = np.array([40,20,20,60,80,60,20,60])

range = np.ptp(data)
print(f"Range: {range}")

iqr = iqr(data)
print(f"IQR: {iqr}")

variance = np.var(data)
print(f"Variance: {variance}")

std = np.std(data)
print(f"Standard Deviation: {std}")

cv = std / np.mean(data)
print(f"Coefficient of Variation: {cv}")
```

```python
data = np.array([40,20,20,60,80,60,20,60])
Q3 = np.quantile(data, 0.75)
Q1 = np.quantile(data, 0.25)
IQR = Q3 - Q1
print(f"Q1 :{Q1}")
print(f"Q3 :{Q3}")
print(f"IQR :{IQR}")
```
