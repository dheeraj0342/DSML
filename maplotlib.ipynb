```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
batter = pd.read_csv('./batter.csv')
batter = batter.head(100).sample(25, random_state=5)
plt.figure(figsize=(10, 10))
plt.scatter(batter['avg'], batter['strike_rate'],s=batter['runs'], alpha=0.5,color='red')
for i in range(batter.shape[0]):
    plt.text(batter['avg'].values[i], batter['strike_rate'].values[i],batter['runs'].values[i])
```

```python
# show bars side by side
batsman_record = pd.read_csv('./batsman_season_record.csv')
plt.figure(figsize=(6,6))
plt.bar(np.arange(batsman_record.shape[0])-.2, batsman_record['2015'], alpha=0.5 ,width=.2, edgecolor='black', linewidth=2,color= 'red')
plt.bar(np.arange(batsman_record.shape[0]), batsman_record['2016'], alpha=0.5 ,width=.2,edgecolor='black', linewidth=2,color= 'orange')
plt.bar(np.arange(batsman_record.shape[0])+.2, batsman_record['2017'], alpha=0.5 ,width=.2,edgecolor='black', linewidth=2,color= 'yellow')
plt.xticks(np.arange(batsman_record.shape[0]), batsman_record['batsman'], rotation=90)
plt.legend(['2015', '2016', '2017'])
```

```python
# stacked bar chart first 2015 then 2016 then 2017
batsman_record = pd.read_csv('./batsman_season_record.csv')
plt.figure(figsize=(6,6))
plt.bar(batsman_record['batsman'], batsman_record['2015'], alpha=0.5 ,width=.5, edgecolor='black', linewidth=2,color= 'red', label='2015')
plt.bar(batsman_record['batsman'], batsman_record['2016'], alpha=0.5 ,width=.5,edgecolor='black', linewidth=2,color= 'orange', bottom=batsman_record['2015'], label='2016')
plt.bar(batsman_record['batsman'], batsman_record['2017'], alpha=0.5 ,width=.5,edgecolor='black', linewidth=2,color= 'yellow', bottom=batsman_record['2015'] + batsman_record['2016'], label='2017')
plt.xticks(rotation=90)
plt.legend(['2015', '2016', '2017'])
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
batter = pd.read_csv('./batter.csv')
fig,ax = plt.subplots( figsize=(6,4))
ax.scatter(batter['avg'], batter['strike_rate'],color='red')
ax.set_xlabel('Average')
ax.set_ylabel('Strike Rate')
ax.set_title('Average vs Strike Rate')
```

```python
fig,ax = plt.subplots(nrows=2, ncols=1, figsize=(6,6))
ax[0].scatter(batter['avg'], batter['strike_rate'],color='red')
ax[0].set_xlabel('Average')
ax[0].set_ylabel('Strike Rate')
ax[0].set_title('Average vs Strike Rate')

ax[1].scatter(batter['avg'], batter['runs'],color='blue')
ax[1].set_xlabel('Average')
ax[1].set_ylabel('Runs')
ax[1].set_title('Average vs Runs')
```

```python
fig,ax = plt.subplots(nrows=2, ncols=2, figsize=(6,6), sharex=True)
ax[0,0].scatter(batter['avg'], batter['strike_rate'],color='red')
ax[0,1].bar(batter['avg'], batter['runs'],color='blue')
ax[1,0].hist(batter['avg'],color='green')
ax[1,1].hist(batter['runs'],color='yellow')
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.subplot(projection='3d')
ax.scatter(batter['avg'], batter['strike_rate'], batter['runs'], color='red')
```

```python
x =np.linspace(-10, 10, 100)
y =np.linspace(-10, 10, 100)
xx, yy = np.meshgrid(x, y)
print(xx,yy)
```

```python
z = np.sin(xx)

fig = plt.figure(figsize=(6,6))

ax = fig.add_subplot( projection='3d')

p=ax.plot_surface(xx, yy, z, cmap='viridis')
fig.colorbar(p)
```

```python


z = xx**2 + yy**2

fig = plt.figure(figsize=(6,6))

ax = fig.add_subplot( projection='3d')

p=ax.plot_surface(xx, yy, z, cmap='viridis')

fig.colorbar(p)
```
