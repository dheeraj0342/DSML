```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

Loading Dataset

```python
df = pd.read_csv('./video_game.csv')
```

checking for null values if present then remove them

```python
df.isnull().sum()
df = df.dropna()
```

Descriptive Statistics

```python
df.describe()
```

Visualization

1.histogram

```python
df['Global_Sales'].hist()
plt.show()
```

2.Box Plots

```python
sns.boxplot(data=df,x='Global_Sales',y='Genre')
```

3.Heatmap

```python

# Select only numeric columns
numeric_cols = df.select_dtypes(include=np.number)

# Compute correlation matrix for numeric columns
corr_matrix = numeric_cols.corr()

# Visualize the correlation matrix with a heatmap
plt.figure(figsize=(10,5))
sns.heatmap(corr_matrix, annot=True)
```

4.Scatter


```python
sns.scatterplot(data=df, x='Global_Sales', y='Critic_Score')
```

Insights


1.Correlation:
If the points on the scatter plot form an upward trend (from left to right), it suggests a positive correlation between 'Global_Sales' and 'Critic_Score'. This means games with higher critic scores tend to have higher global sales. If the points form a downward trend, it suggests a negative correlation, meaning games with lower critic scores tend to have higher global sales. If the points are randomly distributed, it suggests no correlation.

2.Outliers:
Any points that lie far from the general trend of the scatter plot can be considered outliers. These could represent games that had unusually high or low sales relative to their critic scores.

3.Distribution of values:
The scatter plot can also give a sense of the distribution of values within each variable. For example, if the points are concentrated in a certain area of the plot, it suggests most games have critic scores and global sales within that range.
